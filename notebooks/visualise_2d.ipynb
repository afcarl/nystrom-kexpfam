# Importing required modules

We begin with importing all modules.

These involve the data distributions mentioned in the paper: **grid**, **ring**, **glass**.
While **grid** and **ring** have a closed from log-pdf (and gradients), the log-pdf of **glass** can only be estimated pointwise via importance sampling after an EP fit (done using http://shogun.ml/, see paper for details).

It also includes estimator interfaces for **full**, **nystrom**, **lite**, and **dae**, each implementing an interface that contains `fit`, `grad`, and `score`.

Optional notebook dependencies (see also README.md for full list, need at least one, otherwise cannot fit any model):
* **full** and **nystrom** require Shogun
* **lite** requires the Python package https://github.com/karlnapf/kernel_exp_family
* **dae** requires tensorflow

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


from nystrom_kexpfam.data_generators.Gaussian import GaussianGridWrapped
from nystrom_kexpfam.data_generators.Ring import Ring
from nystrom_kexpfam.data_generators.Dataset import GlassPosteriorDataset

from nystrom_kexpfam.log import logger
from nystrom_kexpfam.visualisation import visualise_array_2d

try:
    from nystrom_kexpfam.estimators.DAE import DAE
except ImportError as e:
    logger.error('Tensorflow needs to be installed to use the DAE model: '+ str(e))

try:
    import shogun as sg
    from nystrom_kexpfam.estimators.SGKEF import SGKEF, SGKEFNy, SGKEFNyFixedL2_1e_5, SGKEFNyDFixedL2_1e_5, SGKEFNyD2FixedL2_1e_5
    sg.get_global_io().set_loglevel(2)
except ImportError as e:
    logger.error('Shogun needs to be installed to use the full and Nystrom model: '+ str(e))

try:
    from nystrom_kexpfam.estimators.KMCLite import KMCLite
except ImportError as e:
    logger.error('kernel_exp_family Python package needs to be installed to use the KMCLite model: '+ str(e))
```

# Generating training and test data

```python
np.random.seed(0)

p = GaussianGridWrapped(D=2, sigma=0.1, N_train=100, N_test=1500)
p = Ring(D=2, sigma=0.1, N_train=500, N_test=5000)

X,X_test = p.sample_train_test()
D = X.shape[1]
```

In order to look at pretty density plots, we define a grid over the data space

```python
res = 50
x_min = np.min(X[:,0])
x_max = np.max(X[:,0])
y_min = np.min(X[:,1])
y_max = np.max(X[:,1])
Xs = np.linspace(x_min,x_max, res)
Ys = np.linspace(y_min,y_max,res)
XX, YY = np.meshgrid(Xs, Ys)
X_grid = np.array([XX.ravel(), YY.ravel()]).T
```

# Fitting the estimator

We fit the estimator and predict the gradients of the log-pdf on the test data.
Pick any of the below.

* `KMCLite` is **lite**, i.e. the lite model with a Gaussian kernel and subsampled training data.
* `DAE` is **dae**, i.e. the denoising autoencoder with a decreasing noise schedule.
* `SGKEF` is **full**, i.e. the full kernel exponential family with a Gaussian kernel.
* `SGKEFNyFixedL2_1e_5` is **nystrom**, i.e. the Nystrom kernel exponential family model with a Gaussian kernel and an additional 1e-5 L2 regulariser.


```python
#est = KMCLite(m = 1000, sigma = 3, lmbda = 0.01)
#est = DAE(100, sigma_noise=2, max_iterations=100, num_noise_levels=3)
#est = SGKEF(sigma=2, lmbda = 0.01)
est = SGKEFNyFixedL2_1e_5(m=100, sigma=2, lmbda = 0.01)

est.fit(X)
grad_test = est.grad(X_test)
grad_grid = est.grad(X_grid)
```

We plot samples and estimated density over the grid.
The heatmap corresponds to the Euclidean norm of the gradient, while the arrows indicate (unnormalised) gradient direction and magnitude.

```python
fig=plt.figure(figsize=(16,8))
grad_grid_norm = np.linalg.norm(grad_grid, axis=1)
visualise_array_2d(Xs, Ys, grad_grid_norm.reshape(len(Ys), len(Xs)).T,
                   samples=X, ax=fig.add_subplot(111))
plt.quiver(XX, YY, grad_grid[:,0], grad_grid[:,1], units='width')
plt.title("Estimated gradient of log-density")
plt.xlim([x_min, x_max])
plt.ylim([y_min, y_max])
```

If the true gradient is available, we can compute the true empirical score matching objective.
What the method `score` of each model returns is the score-matching score estimate that doesn't require true gradients, but only training data and diagonal of the model Hessian.
The two are the same up to a constant.

```python
try:
    p.grad_multiple(X_grid)
    true_grad_available = True
except NotImplementedError:
    pass

if true_grad_available:
    grad_test_true = p.grad_multiple(X_test)
    grad_grid_true = p.grad_multiple(X_grid)

    fig=plt.figure(figsize=(16,8))
    grad_grid_true_norm = np.linalg.norm(grad_grid_true, axis=1)
    visualise_array_2d(Xs, Ys, grad_grid_true_norm.reshape(len(Ys), len(Xs)).T,
                       samples=X, ax=fig.add_subplot(111))
    plt.quiver(XX, YY, grad_grid_true[:,0], grad_grid_true[:,1], units='width')
    plt.title("True gradient of log-density")
    plt.xlim([x_min, x_max])
    plt.ylim([y_min, y_max])

    score = np.mean(np.sum((grad_test - grad_test_true)**2, axis=1))
    print "True empirical score of gradient estimator:", score

print "Empirical score of gradient estimator (true score up to constant):", est.score(X_test)
```

The results in the paper were obtained by careful cross-validation (exhaustive grid search) of the score results, and taking the best performing setting for each dimension and method.

Note that while the true score matching loss (average L2 error of estimated gradients) requires the true gradients of the underlying density, the call `est.score(X_test)` does only require test data and first and second gradients of the model. 

# HMC trajectories

We now compute HMC trajectories using the estimated gradients, and compute the average acceptance rate using the log-pdf of the **true** underlying density.

Note that for the Glass dataset, the dataset is a thinned version of a long MCMC chain on the marginal posterior over hyperparameters of a GP classifier, see "Gradient-free Hamiltonian Monte Carlo with efficient kernel exponential families", https://arxiv.org/abs/1506.02564.
The dataset is provided in this repository, but could also easily be generated using any MCMC method using the `log_pdf` implementation of the `GlassPosterior`, see `scripts/mcmc_glass_ground_truth.py` for a minimal script to do that.

The `log_pdf` method itself estimates the (unknown) log-pdf pointwise using importance sampling and expectation propagation in Shogun.
It is important that the log-pdf used to generate the MCMC samples is the same as the one used to evaluate the HMC acceptance (GP and inference parameters need to match).

As each log-pdf evaluation takes a few seconds, computing the averange acceptance probability along 100 HMC steps is slow.
For visualising trajectories quicker, try the **grid** dataset.

```python
from nystrom_kexpfam.hamiltonian import compute_avg_acceptance
```

```python
# N_train for datasets is the (random) fraction of data to use for training
#p = GlassPosteriorDataset("../data/gp_glass_sketch_D=9_N=2000_thinned.npy", N_train=0.05)
p = GaussianGridWrapped(D=2, sigma=0.1, N_train=100, N_test=1500)

X,X_test = p.sample_train_test()
est.fit(X)
```

We set the HMC parameters (as in the paper).
The momentum is an isotropic Gaussian with standard deviation as specified.

```python
sigma_p=1.0 # momentum std deviation
step_size=0.1
num_steps=100
```

We start the trajectory from a random point of the dataset.

```python
q0 = X[np.random.randint(0, len(X))]
```

We hand the estimator's `grad` method to the HMC integrator, and use the **target's** (!= the estimator's) `log_pdf` method to evaluate the acceptance probabilities of the HMC trajectory.

```python
logq = p.log_pdf
dlogq = lambda x : est.grad(np.atleast_2d(x)).ravel() # dlogq is expected to work on 1d arrays
```

```python
# optionally plot HMC trajectory (projected to 2 dimensions)
plot = True
plot_i, plot_j = 0,1
fig=plt.figure(figsize=(16,8))


plot_only_trajectory = (X.shape[1]!=2) # only plot density if space is 2D, otherwise only projected samples

acc_mean = compute_avg_acceptance(q0, logq, dlogq, sigma_p, num_steps, step_size,
                       plot=True, X=X, est=est, plot_i=plot_i, plot_j=plot_j, ax=fig.add_subplot(111),
                       plot_only_trajectory=plot_only_trajectory)

print "Average acceptance rate along trajectory: %.2f" % acc_mean
```
